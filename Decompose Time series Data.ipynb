```python
from random import randrange
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
series = [i+randrange(10) for i in range(1,100)]
result = seasonal_decompose(series, model='additive', freq=1)
result.plot()
plt.show()
```

```python
'''Because the time series was contrived
and was provided as an array of numbers, we must specify the frequency of the observations
(the freq=1 argument). If a Pandas Series object is provided, this argument is not required.'''
```

```python
# multiplicative decompose a contrived multiplicative time series
```

```python
series1 = [i**2.0 for i in range(1,100)]
result1 = seasonal_decompose(series1, model='multiplicative', freq=1)
result1.plot()
plt.show()
```
